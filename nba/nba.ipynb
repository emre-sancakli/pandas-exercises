```python
import pandas as pd
```

```python
df = pd.read_csv("all_seasons.csv")
```

```python
df.head(10)
```

```python
# kayıt sayısı
len(df.index)
```

```python
# yaş ortalaması
df["age"].mean()
```

```python
# en yüksek yaş
df["age"].max()
```

```python
# en yüksek yaşlı oyuncu
df[df["age"] == df["age"].max()]["player_name"].iloc[0]
# iloc[0]'ı sadece ismi almak için kullandık
```

```python
result = df[(df["age"] >= 20) & (df["age"] <= 25)][["player_name","team_abbreviation","age"]].sort_values("age", ascending = False)
result
```

```python
#Elliot Perry'nin oynadığı takımlar ?
result = df[df["player_name"] == "Elliot Perry"]["team_abbreviation"]
result
```

```python
# takımlara göre ortalama player bilgileri
result = df.groupby("team_abbreviation").mean()
result
```

```python
# takımlara göre ortalama player_height bilgileri
result = df.groupby("team_abbreviation").mean()["player_height"]
result
```

```python
# kaç farklı takım var?
result = len(df.groupby("team_abbreviation"))
result
```

```python
# kaç farklı takım var?
result = df["team_abbreviation"].nunique() # num of uniques
result
```

```python
# ismi içerisinde 'and' geçen kayıtları bul
result = df[df["player_name"].str.contains("and")]
result
```
