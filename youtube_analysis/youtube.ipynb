```python
# secili kolonlarƒ± kaldƒ±rƒ±rƒ±z
df = df.drop(["thumbnail_link","comments_disabled","ratings_disabled","video_error_or_removed","description","trending_date"], axis=1)
df
```

```python
# en cok goruntulenen video
result = df[ (df["views"].max()) == df["views"]]["title"].iloc[0]
result
```

```python
# en az izlenen video
result = df[ (df["views"].min()) == df["views"]]["title"].iloc[0]
result
```

```python
# top 10 g√∂r√ºnt√ºlenen,
result = df.sort_values("views", ascending=False)[["title","views"]].head(10)
result
```

```python
#kategoriye g√∂re beƒüeni ort
result = df.groupby("category_id").mean().sort_values("likes")["likes"]
result
```

```python
#kategoriye g√∂re yorum sayƒ±larƒ± yukard a≈üaƒüƒ±
result = df.groupby("category_id").sum().sort_values("comment_count",ascending=False)["comment_count"]
result
```

```python
#her kategoride kac video var?
result = df["category_id"].value_counts()
result
```

```python
# her videonun title uzunlugu bilgisini yeni bir kolonda g√∂ster
df["title_len"] = df["title"].apply(len)
df
```

```python
# her video icin kullanƒ±lan tag sayƒ±sƒ±nƒ± yeni kolonda goster
df["tag_count"] = df["tags"].apply(lambda x: len(x.split('|')))
df
```

```python
# en pop√ºler videolari listele(like/dislike oranƒ±na g√∂re)

def like_dislike_oranla(dataset):
    likesList = list(dataset["likes"])
    dislikesList = list(dataset["dislikes"])
    
    liste = list(zip(likesList,dislikesList))
    
    oranListesi = []
    
    for like,dislike in liste:
        if(like+dislike) == 0:
            oranListesi.append(0)
        else:
            oranListesi.append(like / (like+dislike))
    

    return oranListesi

df["begeni_orani"] = like_dislike_oranla(df)
df.sort_values("begeni_orani",ascending=False)[["title","likes","dislikes","begeni_orani"]]
```
